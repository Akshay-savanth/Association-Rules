```python
import numpy as np
import pandas as pd
df=pd.read_excel("groceries.xls",sep=',')
df=df.replace(np.nan,'',regex=True)
df
```

```python
df_transactions=[]
for i in range(0,9834):
    df_transactions.append([str(df.values[i,u])for u in range(1,31)])
```

```python
grocery_series=pd.DataFrame(pd.Series(df_transactions))
grocery_series
grocery_series.columns=["business"]
grocery_series
```

```python
X=grocery_series['business'].str.join(sep=',').str.get_dummies(sep=',')
X
```

```python
from mlxtend.frequent_patterns import apriori,association_rules
frequent_itemsets=apriori(X,min_support=0.010,max_len=3,use_colnames=True)
frequent_itemsets
```

```python
frequent_itemsets.sort_values('support',ascending=True,inplace=True)
frequent_itemsets.sort_values
```

```python
rules=association_rules(frequent_itemsets,metric="lift",min_threshold=2)
rules.head(30)
rules.sort_values("lift",ascending=False).head(20)
```
