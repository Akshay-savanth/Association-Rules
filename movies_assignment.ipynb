```python
import pandas as pd
movies=pd.read_csv("my_movies.csv")
movies
```

```python
movies.isnull().sum()
```

```python
movies=movies.drop(['V1'],axis=1)
movies=movies.drop(['V2'],axis=1)
movies=movies.drop(['V3'],axis=1)
movies=movies.drop(['V4'],axis=1)
movies=movies.drop(['V5'],axis=1)
movies
```

```python
X=movies
X
```

```python
!pip install mlxtend
```

```python
from mlxtend.frequent_patterns import apriori,association_rules
frequent_itemsets=apriori(X,min_support=0.015,max_len=3,use_colnames=True)
frequent_itemsets
```

```python
frequent_itemsets.sort_values('support',ascending = False,inplace=True)
frequent_itemsets.sort_values
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=2)
rules.head(20)
rules.sort_values('lift',ascending = False).head(20)
```
